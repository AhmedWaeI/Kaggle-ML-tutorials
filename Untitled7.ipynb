```python
import billboard
import pandas as pd
char=[]
```

```python
year= range(2022,2017,-1)
for y in year:
    char.extend(billboard.ChartData('hot-100-songs',year=y))
```

```python
data = {'title': [], 'artist': []}


for entry in char:
    data['title'].append(entry.title)
    data['artist'].append(entry.artist)
df = pd.DataFrame(data)
```

```python
df['Popularity']=1
```

```python
for artist in df['artist'].unique():
    results = sp.search(q='artist:' + artist, type='track', limit=10)
    songs = results['tracks']['items']
    all_songs = [song['name'] for song in songs if '2018-01-01' <= song['album']['release_date'] <= '2022-12-31']
    all_songs=set(all_songs)
    filteredsongs=[]
    for song in all_songs:
        if song not in df['title'].values:
            filteredsongs.append(song)
    for song in filteredsongs:
        row={'title': song, 'artist':artist, 'Popularity':0}
        df=df.append(row,ignore_index=True)
```
