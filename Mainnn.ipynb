```python
import yt_dlp
import os

def download_song(song_name, artist_name):
    search_query = f'{song_name} {artist_name} audio'
    output_file = os.path.join("C:/Users/slayer/Desktop/Newfolder", f'{song_name} {artist_name}.mp3')

    # Check if the file already exists
    if os.path.exists(output_file):
        print(f'{song_name} {artist_name} already downloaded. Skipping...')
        return

    ydl_opts = {
        'default_search': f'ytsearch:{search_query}',
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'postprocessor_args': [
            '-ar', '44100'
        ],
        'prefer_ffmpeg': True,
        'noplaylist': True,
        'nocheckcertificate': True,
        'quiet': True,
        'outtmpl': output_file,
        'audioformat': "mp3",
        'ffmpeg_location': 'C:/Users/slayer/Desktop/New folder/ffmpeg-6.0-full_build/bin'
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([search_query])
        print(f'{song_name} {artist_name} downloaded successfully.')
    except Exception as e:
        print(f'Error occurred while downloading {song_name} {artist_name}: {str(e)}')
```

```python
for index, row in df.iterrows():
    download_song(song_name=row.title,artist_name=row.artist)
```

```python
import pychorus as pc
import soundfile as sf

def extract_chorus(audio_file, output_file):
    try:
        # Read the audio file
        data, samplerate = sf.read(audio_file)

        # Get the chromagram
        chroma, _, _, song_length_sec = pc.create_chroma(audio_file)

        # Set the clip length (in seconds)
        clip_length = 15

        # Find the chorus section
        chorus_start = pc.find_chorus(chroma, samplerate, song_length_sec, clip_length)

        if chorus_start is None:
            print("No chorus detected.")
            return False

        # Set the duration of the chorus (in seconds)
        chorus_duration = 15

        # Calculate start and end time of chorus
        start_time = int(chorus_start * samplerate)
        end_time = start_time + int(chorus_duration * samplerate)

        # Extract the chorus section
        chorus = data[start_time:end_time]

        # Write chorus to WAV file
        sf.write(output_file, chorus, samplerate)
        print(f"Chorus extracted and saved as {output_file}.")
        return True

    except Exception as e:
        print(f"Error occurred for {audio_file}: {e}")
        return False
```
